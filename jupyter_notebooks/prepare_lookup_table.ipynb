```python
import pandas as pd 
import numpy as np
```

# Import Players 2022 Data and the Dataset

```python
teams_features = pd.read_csv("../data/world_cup_2022_data/teams_features.csv")
teams_features.head()
```

```python
dataset = pd.read_csv("../data/dataset/dataset.csv")
dataset.head()
```

```python
lookup_table = pd.DataFrame()
done_countries = []
for country1 in teams_features["country"].unique():
    print(country1)
    features_coun1 = teams_features[teams_features["country"] == country1].reset_index(drop=True)
    features_coun1_cols = [col + "_1st" for col in features_coun1.columns]
    features_coun1.columns = features_coun1_cols
    for country2 in teams_features["country"].unique():
        if country2 not in done_countries and country1 != country2:
            features_coun2 = teams_features[teams_features["country"] == country2].reset_index(drop=True)
            features_coun2_cols = [col + "_2nd" for col in features_coun2.columns]
            features_coun2.columns = features_coun2_cols
            concated_features = pd.concat([features_coun1, features_coun2], ignore_index=True,axis=1)
            concated_features.columns = features_coun1_cols + features_coun2_cols
            lookup_table = pd.concat([lookup_table, concated_features], ignore_index=True, axis=0)
            lookup_table.columns = features_coun1_cols + features_coun2_cols


    done_countries.append(country1)

lookup_table.rename(columns={"country_1st":"1st_team", "country_2nd":"2nd_team"}, inplace=True)
```

```python
lookup_table
```

```python
desired_cols_ordere = dataset.drop(columns=["year", "net_score"]).columns
```

```python
lookup_table_oredered = lookup_table[desired_cols_ordere]
```

```python
lookup_table_oredered
```

```python
assert all(lookup_table_oredered.columns == dataset.drop(columns=["year", "net_score"]).columns)
```

```python
assert len(desired_cols_ordere) == len(lookup_table.columns)
assert len(lookup_table.columns) == len(lookup_table_oredered.columns)
assert all(lookup_table_oredered.columns == dataset.drop(columns=["year", "net_score"]).columns)
```

```python
lookup_table_oredered.to_csv("../data/dataset/lookup_table.csv", index=False)
```
